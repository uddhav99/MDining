```python
import numpy as np
import pandas as pd
menu = pd.read_csv("./data/menu.csv", index_col=0)
dining = pd.read_csv("./data/dining.csv")
```

```python
menu.head()
```

```python
menu.columns = ['serving_size', 'calories', 'total fat (g)', 'sodium (g)', 'carbs (g)', 'fiber (g)', 'sugar (g)', 'protein (g)', 'vitaminA', 'vitaminC', 'calcium', 'iron']
```

```python
# changing from object to str or float and replacing g with 0
menu["carbs (g)"] = menu["carbs (g)"].astype(str)
menu["carbs (g)"] = menu["carbs (g)"].str.replace('g', '')
menu["carbs (g)"] = menu["carbs (g)"].str.replace('%', '0')
menu["carbs (g)"] = menu["carbs (g)"].astype('float')
```

```python
menu.head()
```

```python
# changing from object to str and replacing units and making units according to header units 
menu["total fat (g)"] = menu["total fat (g)"].astype(str)
menu["total fat (g)"] = menu["total fat (g)"].str.replace('g', '')
menu["total fat (g)"] = menu["total fat (g)"].astype('float')
menu["sodium (g)"] = menu["sodium (g)"].astype(str)
menu["sodium (g)"] = menu["sodium (g)"].str.replace('m', '')
menu["sodium (g)"] = menu["sodium (g)"].str.replace('g', '')
menu["sodium (g)"] = menu["sodium (g)"].str.replace('%', '0')
menu["sodium (g)"] = menu["sodium (g)"].astype('float')
menu["sodium (g)"] = menu["sodium (g)"]/1000
menu["fiber (g)"] = menu["fiber (g)"].astype(str)
menu["fiber (g)"] = menu["fiber (g)"].str.replace('g', '')
menu["fiber (g)"] = menu["fiber (g)"].astype('float')
menu["sugar (g)"] = menu["sugar (g)"].astype(str)
menu["sugar (g)"] = menu["sugar (g)"].str.replace('g', '')
menu["sugar (g)"] = menu["sugar (g)"].str.replace('%', '0')
menu["sugar (g)"] = menu["sugar (g)"].astype('float')
menu["protein (g)"] = menu["protein (g)"].astype(str)
menu["protein (g)"] = menu["protein (g)"].str.replace('g', '')
menu["protein (g)"] = menu["protein (g)"].astype('float')
```

```python
menu.head()
```

```python
# for vitamins and iron to remove %
def rmp(string): #function to replace XX% with XX
    if string == "NaN":
        newstring = ""
    newstr = string.replace("%", "")
    return newstr
```

```python
# creating new columns with standard values for vitamin and iron
menu[["vitaminA", "vitaminC", "calcium", "iron"]] = menu[["vitaminA", "vitaminC", "calcium", "iron"]].astype(str)
menu["vitaminA_mcg"] = menu["vitaminA"].apply(rmp)
menu["vitaminC_mg"] = menu["vitaminC"].apply(rmp)
menu["calcium_mg"] = menu["calcium"].apply(rmp)
menu["iron_mg"] = menu["iron"].apply(rmp)
menu.head()
```

```python
# converting all vitamin, iron items to their standard values 
menu[menu["vitaminA_mcg"]==""].head() 
menu.replace(r'^\s*$', np.nan, regex=True, inplace = True) #replace blank space with NaN
menu[["vitaminA_mcg", "vitaminC_mg", "calcium_mg", "iron_mg"]] = menu[["vitaminA_mcg", "vitaminC_mg", "calcium_mg", "iron_mg"]].astype(float)
menu["vitaminA_mcg"] = 900*(menu["vitaminA_mcg"]/100)
menu["vitaminC_mg"] = 90*(menu["vitaminC_mg"]/100)
menu["calcium_mg"] = 1300*(menu["calcium_mg"]/100)
menu["iron_mg"] = 18*(menu["iron_mg"]/100)
menu.head()
```

```python
series_temp_grams = pd.Series(np.asarray((len(menu['serving_size']))))
series_temp_serve = pd.Series(np.zeros(len(menu['serving_size']))) 

indexlist = menu['serving_size'].loc[menu['serving_size'].notna()].index
i = 0
for string in menu['serving_size'].loc[menu['serving_size'].notna()]:
    
    series_temp_grams.loc[indexlist[i]] = int(string.split()[-1][1:-2])
    series_temp_serve.loc[indexlist[i]] = ' '.join(string.split()[0:-1])
    i = i+1
```

```python
menu.insert(1,'serving_size_(g)',series_temp_grams)
menu.insert(1,'serving_size_units',series_temp_serve)
```

```python
menu.head()
```

```python
menu["serving_size_(g)"] = menu["serving_size_(g)"].astype(float)
menu["serving_size_units"] = menu["serving_size_units"].astype(str)
```

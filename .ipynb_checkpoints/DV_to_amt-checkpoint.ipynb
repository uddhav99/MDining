```python
import pandas as pd
import numpy as np
```

```python
dining = pd.read_csv("/Users/rubyhickman/Documents/01_michigan/mdst/MDining/data/menu.csv")
```

```python
dining.head()
dining.shape
```

```python
def rmp(string): #function to replace XX% with XX
    if string == "NaN":
        newstring = ""
    newstr = string.replace("%", "")
    return newstr
```

```python
dining[["vitaminA", "vitaminC", "calcium", "iron"]] = dining[["vitaminA", "vitaminC", "calcium", "iron"]].astype(str)
dining["vitaminA_mcg"] = dining["vitaminA"].apply(rmp)
dining["vitaminC_mg"] = dining["vitaminC"].apply(rmp)
dining["calcium_mg"] = dining["calcium"].apply(rmp)
dining["iron_mg"] = dining["iron"].apply(rmp)
dining.head()
```

```python
dining[dining["vitaminA_mcg"]==""].head() 
dining.replace(r'^\s*$', np.nan, regex=True, inplace = True) #replace blank space with NaN
dining.head()
```

```python
dining[["vitaminA_mcg", "vitaminC_mg", "calcium_mg", "iron_mg"]] = dining[["vitaminA_mcg", "vitaminC_mg", "calcium_mg", "iron_mg"]].astype(float)
```

```python
dining["vitaminA_mcg"] = 900*(dining["vitaminA_mcg"]/100)
dining["vitaminC_mg"] = 90*(dining["vitaminC_mg"]/100)
dining["calcium_mg"] = 1300*(dining["calcium_mg"]/100)
dining["iron_mg"] = 18*(dining["iron_mg"]/100)
```

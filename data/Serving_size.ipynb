```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./menu.csv", index_col = 0)
```

```python
df['serving_size'].iloc[0]
```

```python
series_temp_grams = pd.Series(np.asarray((len(df['serving_size']))))
series_temp_serve = pd.Series(np.zeros(len(df['serving_size']))) 

indexlist = df['serving_size'].loc[df['serving_size'].notna()].index
i = 0
for string in df['serving_size'].loc[df['serving_size'].notna()]:
    
    series_temp_grams.loc[indexlist[i]] = int(string.split()[-1][1:-2])
    series_temp_serve.loc[indexlist[i]] = ' '.join(string.split()[0:-1])
    print(string)
    i = i+1
    
    
```

```python
df.insert(1,'serving_size_(g)',series_temp_grams) 
```

```python
df.insert(1,'serving_size_units',series_temp_serve) 
```

```python
menu.columns = ['serving_size', 'calories', 'total fat (g)', 'sodium (g)', 'carbs (g)', 'fiber (g)', 'sugar (g)', 'protein (g)', 'vitaminA', 'vitaminC', 'calcium', 'iron']
```

```python
menu.head()
```

```python
menu["carbs (g)"] = menu["carbs (g)"].astype(str)
```

```python
menu["carbs (g)"] = menu["carbs (g)"].str.replace('g', '')
menu["carbs (g)"] = menu["carbs (g)"].str.replace('%', '0')
```

```python
menu["carbs (g)"] = menu["carbs (g)"].astype('float')
```

```python
import datetime

import pandas as pd
import requests
from bs4 import BeautifulSoup

menu_items = dict()
meal_items = []

dining_halls = ["bursley"] #list of dining halls
dates = pd.date_range(datetime.datetime(2020, 2, 4), datetime.datetime.today()).tolist()
for date in dates:
    for dh in dining_halls:
        r = requests.get("https://dining.umich.edu/menus-locations/dining-halls/{}/?menuDate={}".format(dh, str(date.date())))
        soup = BeautifulSoup(r.text, "html5lib")
        print("Parsing {} {}".format(dh, str(date.date())))
        items = soup.select("ul.items > li")
        for item in items:
            name = item.select('.item-name')[0].text
            # add {item, dining hall, meal, date} to meal_items
            meal_items.append({
                "date": str(date.date()),
                "dining_hall": dh,
                "meal": None,
                "item": name
            })
            if name not in items:
                # add {item, nutrition, other attrs} info to dict
                menu_items[name] = dict()
                for tr in item.select(".nutrition-wrapper tr"):
                    if "Serving Size" in tr.text:
                        menu_items[name]["serving_size"] = " ".join(tr.text.split()[2:])
                    if "Calories" in tr.text:
                        menu_items[name]["calories"] = "".join(tr.text.split()[1:])
                        
```
